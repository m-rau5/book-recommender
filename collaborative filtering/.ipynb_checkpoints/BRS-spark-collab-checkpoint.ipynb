# Setup

```python
from pyspark.sql import SparkSession 
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.recommendation import ALS, ALSModel
from pyspark.ml.tuning import TrainValidationSplit,CrossValidator, ParamGridBuilder
from pyspark.sql.functions import col, explode, when
```

__Pyspark setup__

```python
import os
import sys

os.environ['PYSPARK_PYTHON'] = sys.executable
os.environ['PYSPARK_DRIVER_PYTHON'] = sys.executable
```

```python
spark = SparkSession.builder \
    .master('local') \
    .appName('BRS-pyspark') \
    .config("spark.driver.memory", "12g") \
    .config("spark.driver.extraJavaOptions", "-Xss8M") \
    .getOrCreate()

spark.conf.set("spark.sql.execution.arrow.pyspark.enabled", "true")
```

```python
spark
```

---

# Loading the datasets

```python
books_df = spark.read.csv("../datasets/clean/filtered_datasets/Final/final_books.csv",header=True,inferSchema=True)
book_ratings = spark.read.csv("../datasets/clean/filtered_datasets/Final/final_ratings.csv",header=True,inferSchema=True)
```

```python
book_ratings.show(2)
```

```python
(training, test) = book_ratings.randomSplit([0.8,0.2], seed=42)
```

## Setting up evaluator, model and paramaters

```python
# Evaluator = RMSE -> sqroot of the avg squared differences between 
#                                   predicted and observed outcomes
evaluator = RegressionEvaluator(metricName="rmse", labelCol="Book-Rating",
                                predictionCol="prediction")
```

```python
als = ALS(userCol="User-ID",itemCol="Book-Id", ratingCol="Book-Rating",
                coldStartStrategy="drop",nonnegative=True)
```

```python
# param_grid -> for fine-tuning of the model -> hyperparams
param_grid = ParamGridBuilder()\
             .addGrid(als.rank,[70])\
             .addGrid(als.maxIter,[140])\
             .addGrid(als.regParam, [0.9])\
             .build()
# hyperparam "rank" - rank of U and P 
# hyperparam "maxIter" - how many times to alternate
# hyperparam "regParam" - regularization param to prevent overfitting
```

We use cross validator to create multiple models and choose best one

```python
cv = CrossValidator(estimator=als,
                    estimatorParamMaps=param_grid,
                    evaluator=evaluator,
                    numFolds=4) 
```

Train the models

```python
model = cv.fit(training)
bestModel = model.bestModel
```

---

# Testing the model

Now we see how the model performs on our test dataset we created

```python
pred = bestModel.transform(test)
rmse = evaluator.evaluate(pred)
rmse
```

```python
print("rank :",bestModel.rank," maxiter :",bestModel._java_obj.parent().getMaxIter(), " regparam :",bestModel._java_obj.parent().getRegParam())
```

#### __Code to plot user reviews for comparison__

```python
import matplotlib.pyplot as plt
import pandas as pd

def plotUserRatings(user_id, book_ratings, book_df):
    # filter book ratings for the user
    ratings = book_ratings.filter(book_ratings["User-ID"] == user_id)
    ratings_with_names = ratings.join(book_df, on="Book-Id", how="inner")
    
    book_names = ratings_with_names.select("Book-Title").collect()
    ratings = ratings_with_names.select("Book-Rating").collect()
    
    # make the spark dataframe into a pandas df for plotting
    df = pd.DataFrame({"Book Title": [row["Book-Title"] for row in book_names],
                       "Rating": [row["Book-Rating"] for row in ratings]})

    # --- plotting code ---
    plt.figure(figsize=(10, 6))
    plt.barh(df["Book Title"], df["Rating"], color="skyblue")
    plt.xlabel("Rating")
    plt.ylabel("Book Title")
    plt.title(f"Books Rated by User {user_id}")
    plt.gca().invert_yaxis()
    plt.show()
```

### Function to get user recommendations from the model

```python
def scale_ratings(predictions):
    # Get the maximum and minimum predicted ratings
    max_rating = predictions.selectExpr("max(`Predicted-Rating`)").collect()[0][0]
    min_rating = 0

    # Scale the ratings to the range [0, 10]
    scaled_predictions = predictions.withColumn("Scaled-Rating", ((col("Predicted-Rating") - min_rating) / (max_rating - min_rating)) * 10)

    return scaled_predictions

def recommendBooks(user_id, model, book_df):
    user_recs = model.recommendForUserSubset(spark.createDataFrame([[user_id]], ["User-ID"]), numItems=10)
    
    # Use "explode" on the array of recommendations to separate rows
    user_recs_exploded = user_recs.select(col("User-ID"), explode("recommendations").alias("recommendation"))
    
    # Get recommended books and ratings + join on their name
    recommended_books = user_recs_exploded.select(col("User-ID"), col("recommendation.Book-Id").alias("Book-Id"), col("recommendation.rating").alias("Predicted-Rating"))
    recommended_books_with_names = recommended_books.join(book_df, on="Book-Id", how="inner")

    # Scale the ratings to the range [0, 10]
    scaled_recommendations = scale_ratings(recommended_books_with_names)

    return scaled_recommendations
```
