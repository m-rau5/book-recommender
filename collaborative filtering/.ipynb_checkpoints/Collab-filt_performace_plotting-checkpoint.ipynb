# Plot of Rank effect on RMSE

```python
import matplotlib.pyplot as plt
# Data
ranks = [20, 50, 90, 130, 150]
rmse_scores = [2.1988, 2.103, 2.1002, 2.0965, 2.102]

# Create the plot
plt.figure(figsize=(7, 5))
plt.plot(ranks, rmse_scores, marker='o', linestyle='-', color='b')
plt.xlabel('Rank')
plt.ylabel('RMSE')
plt.title('Effect of Rank on RMSE')
plt.grid(True)
plt.xticks(ranks)
plt.yticks(sorted(rmse_scores))

# Annotate points
for i, txt in enumerate(rmse_scores):
    plt.annotate(f"{txt:.4f}", (ranks[i], rmse_scores[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Show the plot
plt.show()
```
