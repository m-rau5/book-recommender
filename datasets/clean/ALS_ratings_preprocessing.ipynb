# Setting up the ratings dataset

We are using two datasets: 
- __merged_books_df__: the main books dataset used, defined in the Data Merging part
- __ratings_large_df__: a large ratings dataset that came bundled with the large books dataset used in the Data Merging part.

Methodology: We will keep all ratings of books present in the books datasaet and deal with any potential errors or problems we might encounter with the resulting dataset

```python
merged_books_df = pd.read_csv("./filtered_datasets/books_merged_filtered.csv")
ratings_large_df = pd.read_csv("source_datasets/book_dataset_large/ratings-large.csv",low_memory=False)
```

We first get all ratings of books present in the merged books dataset

```python
ratings_present_df = ratings_large_df[ratings_large_df['ISBN'].isin(merged_books_df['ISBN'])]
```

```python
print(ratings_present_df.shape) # see the amount of reviews, from 40539 to 40121
print(ratings_present_df['ISBN'].nunique()) # see if the books review match the books in dataset
```

```python
ratings_present_df = ratings_present_df.drop_duplicates(subset=['ISBN', 'User-ID'])

print("Number of unique ratings after removing duplicates:", ratings_present_df.shape[0])
ratings_present_df.head(1)
```

## Problem No. 1 -> Unnecesary data + data formatting

#### Let's look at the reviews:

Review distribution:

```python
import matplotlib.pyplot as plt

# Plot histogram of ratings
plt.figure(figsize=(7, 4))
plt.hist(ratings_present_df['Book-Rating'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Book Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

We can clearly see above the amount of reviews with value 0. These are inherently _useless_, since they represent a lack of opinion/review on a book the user has read.

__Solution:__ we remove 0's from the equation

```python
ratings_0 = ratings_present_df[ratings_present_df["Book-Rating"] ==0]
ratings_1to10 = ratings_present_df[ratings_present_df["Book-Rating"] !=0]
```

### __View ratings distribution -> 0's removed__

```python
import matplotlib.pyplot as plt

# Plot histogram of ratings
plt.figure(figsize=(7, 4))
plt.hist(ratings_1to10['Book-Rating'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Book Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

print("The shape of the ratings is: " + str(ratings_1to10.shape))
```

## Problem No. 2 -> Distribution of data

Let's view the data of the users that left reviews

```python
user_books_count = ratings_1to10.groupby('User-ID')['ISBN'].nunique().reset_index()
user_books_count.columns = ['user_id', 'books_reviewed_count']
print(user_books_count.head(3))
```

We can see how many reviews each user has left. let's make it easier to see:

#### We can look at the distribution of revews for the users

```python
filtered_user_books_count = user_books_count.groupby('books_reviewed_count').size().reset_index(name='num_users')

filtered_user_books_count = filtered_user_books_count.sort_values(by='books_reviewed_count')

review_distribution = filtered_user_books_count.head(10).to_string(index=False)
print(review_distribution)
```

__Problem:__ ! The users with less reviews skew the dataset. !



```python
# Plotting the distribution of user reviews
plt.figure(figsize=(10, 6))
plt.bar(filtered_user_books_count['books_reviewed_count'], filtered_user_books_count['num_users'], color='skyblue')

# Set labels and title
plt.xlabel('Number of Books Reviewed')
plt.ylabel('Number of Users')
plt.title('Distribution of User Reviews')

max_books_reviewed = 20
plt.xlim(0, max_books_reviewed)  # Set x-axis limits from 0 to max_books_reviewed

# Show plot
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add horizontal grid lines
plt.xticks(range(1, max_books_reviewed + 1))  # Show ticks from 0 to max_books_reviewed
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```

You can observe in the figure above, the users with few books reviewed skew the dataset heavily.

__Solution:__ For the purpose of collaborative filtering, we can remove users with less than 3 reviews since they won't have that large of an influence on the results.

```python
filtered_user_books_count = user_books_count[user_books_count['books_reviewed_count'] > 2]
books_reviewed_freq = filtered_user_books_count['books_reviewed_count'].value_counts().sort_index()
```

Let's plot the results:

```python
plt.figure(figsize=(7, 5))
plt.bar(books_reviewed_freq.index, books_reviewed_freq.values, color='skyblue')
plt.xlabel('Number of Books Reviewed')
plt.ylabel('Number of Users')
plt.title('Distribution of Books Reviewed per User (Excluding Users with 2 Reviews or under)')

max_books_reviewed = 21
plt.xlim(2.5, max_books_reviewed) 

# Show plot
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.xticks(range(3, max_books_reviewed + 1))
plt.tight_layout()
plt.show()
```

The data distribution looks much better now.

Let's look at the shape of the dataset before and after

```python
print("before",ratings_1to10.shape)
user_review_counts = ratings_1to10['User-ID'].value_counts()
active_users = user_review_counts[user_review_counts >= 3].index
ratings_1to10_over3 = ratings_1to10[ratings_1to10['User-ID'].isin(active_users)]
print("after",ratings_1to10_over3.shape)
```

__Consequence:__ There are way fewer users, however, all of them will have a profile that will be usefull for the collaborative filtering part.
